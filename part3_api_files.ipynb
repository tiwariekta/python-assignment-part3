```python
# Task 1 — File Read & Write Basics

# Part A — Write: Create a Python script that writes the following student notes to a file called python_notes.txt:
notes = [
    "Topic 1: Variables store data. Python is dynamically typed.",
    "Topic 2: Lists are ordered and mutable.",
    "Topic 3: Dictionaries store key-value pairs.",
    "Topic 4: Loops automate repetitive tasks.",
    "Topic 5: Exception handling prevents crashes."
]

# 1. Write the above five lines to the file using write mode ('w'). Remember to include encoding="utf-8".
with open("python_notes.txt", "w", encoding="utf-8") as file:
    for line in notes:
        file.write(line + "\n")

print("File written successfully!")

# 2. Append two more lines of your own to the same file using append mode ('a').
additional_notes = [
    "Topic 6: Functions/Definition help in code reusability.",
    "Topic 7: File handling allows persistent storage."
]

with open("python_notes.txt", "a", encoding="utf-8") as file:
    for line in additional_notes:
        file.write(line + "\n")

print("Lines appended successfully!")
```

```python
# Part B — Read
# 1. Read the file back and print each line numbered, stripping the trailing newline character (\n) from each line
with open("python_notes.txt", "r", encoding="utf-8") as file:
    lines = file.readlines()

print("\nReading python_notes.txt file Content:\n")

for i, line in enumerate(lines, start=1):
    print(f"{i}. {line.strip()}")

# 2. Count and print the total number of lines in the file.
total_lines = len(lines)
print(f"\nTotal number of lines: {total_lines}")

# 3. Ask the user for a keyword and print all lines that contain that keyword (case-insensitive). If no match is found, print a helpful message.
while True:
    keyword = input("\nEnter keyword to search (or type 'exit' to stop): ").lower()
    
    if keyword in ["exit", "stop", "quit"]:
        print("\n*** Search ended ***")
        break
    
    if not keyword:
        print("Please enter a valid keyword.")
        continue
    
    # Print user input
    print(f"\nYou searched for: '{keyword}'")
    
    matches = []
    
    for line in lines:
        if keyword in line.lower():
            matches.append(line.strip())
    
    if matches:
        print("\nMatching lines:")
        for line in matches:
            print(line)
    else:
        print("No matching lines found.")   
```

```python
# Task 2 — API Integration
# Step 1 — Fetch and Display Products:

import requests

# Make a GET request to fetch 20 products:
url = "https://dummyjson.com/products?limit=20"

response = requests.get(url)

# Parse the response to  JSON
data = response.json()

products = data["products"]

# Print a formatted table
print("ID  | Title                          | Category      | Price    | Rating")
print("-" * 75)

for p in products:
    print(f"{p['id']:<3} | {p['title']:<30} | {p['category']:<13} | ${p['price']:<8} | {p['rating']}")
```

```python
# Step 2 — Filter and Sort:
# From the 20 fetched products, filter only those with a rating ≥ 4.5
filtered = [p for p in products if p["rating"] >= 4.5]

# Sort the filtered list by price in descending order
filtered_sorted = sorted(filtered, key=lambda x: x["price"], reverse=True)

# Print the filtered and sorted list
print("\nFiltered & Sorted Products:\n")
print("ID  | Title                                      | Category      | Price    | Rating")    
print("-" * 85)
for p in filtered_sorted:
    print(f"{p['id']:<3} | {p['title']:<42} | {p['category']:<13} | ${p['price']:<8} | {p['rating']}")
```

```python
# Step 3 — Search by Category
# Make a GET request to fetch all products in the laptops category:

laptop_url = "https://dummyjson.com/products/category/laptops"

response = requests.get(laptop_url)
laptops_data = response.json()

laptops = laptops_data["products"]

# Print the name and price of each laptop found.
print("\nLaptops:\n")

print("Laptop Name                                | Price")    
print("-" * 55)

for p in laptops:
    print(f"{p['title']:<42} | ${p['price']:<8}")

```

```python
# Task 3 — Exception Handling

# Part A — Guarded Calculator:

def safe_divide(a, b):
    try:
        return a / b
    
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    
    except TypeError:
        return "Error: Invalid input types"
    
# Test it with: safe_divide(10, 2), safe_divide(10, 0), safe_divide("ten", 2). Print each result.
print(safe_divide(10, 2))  
print(safe_divide(10, 0))   
print(safe_divide("ten", 2))  
```

```python
# Part B — Guarded File Reader:

def read_file_safe(filename):
    try:
        with open(filename, "r", encoding="utf-8") as file:
            content = file.read()
            return content
    
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
    
    finally:
        print("File operation attempt complete.")

# Test it with "python_notes.txt" (should succeed) and "ghost_file.txt" (should fail gracefully).
print(read_file_safe("python_notes.txt"))
print(read_file_safe("ghost_file.txt")) 
```

```python
# Part C — Robust API Calls:

import requests

url = "https://dummyjson.com/products?limit=20"

try:
    response = requests.get(url, timeout=5)
    data = response.json()
    print("API call successful")

except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")

except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")

except Exception as e:
    print(f"Unexpected error: {e}")
```

```python
#Part D — Input Validation Loop:

while True:

    # 1. Asks the user: "Enter a product ID to look up (1–100), or 'quit' to exit: "
    user_input = input("\nEnter a product ID (1-100) or 'quit' to exit: ")

    # Print User input
    print(f"You entered: '{user_input}'")
    
    # 5. Continues until the user types quit
    if user_input.lower() == "quit":
        print("Exiting...")
        break
    
    # 2. Validates that the input is an integer in the range 1–100. If not, print a warning and ask again do not make an API call with invalid input.
    # Validate integer
    if not user_input.isdigit():
        print("Please enter a valid number.")
        continue
    
    product_id = int(user_input)
    
    # Validate range
    if product_id < 1 or product_id > 100:
        print("Please enter a number between 1 and 100.")
        continue
    
    url = f"https://dummyjson.com/products/{product_id}"
    
    try:
        # 3. Makes a GET request to https://dummyjson.com/products/{id}
        response = requests.get(url, timeout=5)
        
        # 4. Checks response.status_code: if it is 404, print "Product not found." If it is 200, print the product's title and price
        if response.status_code == 404:
            print("Product not found.")
        
        elif response.status_code == 200:
            product = response.json()
            print(f" {product['title']} — ${product['price']}")
        
        else:
            print(f" Unexpected status code: {response.status_code}")
    
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please check your internet.")
    
    except requests.exceptions.Timeout:
        print("Request timed out. Try again later.")
    
    except Exception as e:
        print(f"Unexpected error: {e}")
```

```python
# Task 4 — Logging to File

from datetime import datetime

def log_error(function_name, error_type, message):
    
    # 2. Use Python's datetime module (datetime.now()) to timestamp each entry.
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    
    log_entry = f"[{timestamp}] ERROR in {function_name}: {error_type} — {message}\n"
    
    # 1. Open error_log.txt in append mode so logs accumulate across runs.
    with open("error_log.txt", "a", encoding="utf-8") as file:
        file.write(log_entry)
```
